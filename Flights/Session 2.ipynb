```python
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("flights.csv")

```

```python
data.shape
```

```python
#remove the column Day of the week
data.drop('DAY_OF_WEEK', axis=1, inplace=True)
data.shape
```

```python
#Rename the column WHEELS_OFF to HAS_WHEELS
data.rename(columns={'WHEELS_OFF': 'HAS_WHEELS'}, inplace=True)
```

```python
data.describe()
```

```python
#slice the dataset into 4 equal chunks
df_split = np.array_split(data, 4)
```

```python
#concatenate
data_new = pd.concat([df_split[0],df_split[1],df_split[2],df_split[3]])

data_new = data_new.reset_index(drop=True)
```

```python
data_new.head(5)
```

```python
#only for Airline AA
data_AA = data_new.loc[data_new['AIRLINE'] == 'AA']
```

```python
data_AA['AIRLINE'].value_counts()
```

```python
#delay < 10, Destination = PBI
data_10_PBI = data_new[(data_new['DEPARTURE_DELAY']<10) & (data_new['DESTINATION_AIRPORT']=='PBI')]
```

```python
data_10_PBI['DESTINATION_AIRPORT'].value_counts()
```

```python
#Impute AIR SYSTEM DELAY with mean
print(data_new['AIR_SYSTEM_DELAY'].isna().sum())
data_new['AIR_SYSTEM_DELAY'] = data_new['AIR_SYSTEM_DELAY'].fillna((data_new['AIR_SYSTEM_DELAY'].mean()))
print(data_new['AIR_SYSTEM_DELAY'].isna().sum())
```

```python
#indicator columns has_A
data_new['has_A'] = np.where(data_new['AIRLINE'].apply(lambda x: 'A' in x), 1, 0)
```

```python
#random Sample of rows in DF
data_sample=data_new.sample(20)
```

```python
#Min max for departure delay
data2=data_new
scaler = MinMaxScaler()
```

```python
data2['DEPARTURE_DELAY'] = data2['DEPARTURE_DELAY'].fillna((data2['DEPARTURE_DELAY'].mean()))
```

```python
data2[['DEPARTURE_DELAY']] = scaler.fit_transform(data2[['DEPARTURE_DELAY']])
```

```python
data2.describe()
```

```python
data2.shape
```

```python
#binarize Origin Airport
data3 = pd.get_dummies(data2['ORIGIN_AIRPORT'])
data3.columns
```

```python
data4 = pd.concat([data2, data3], axis=1)
```

```python
data4.shape
```

```python
data4.to_csv('data_final.csv')
```
